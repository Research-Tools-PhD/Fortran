# Sorting Algorithm Performance Comparison

 To compare the performance of different *sorting algorithms*, we measured the time it took to sort arrays of various sizes and configurations. The sorting algorithms we tested are:
>Simple Sort

>Bubble Sort 

>Selection Sort

> Insertion Sort

> Quick Sort

> Merge Sort

>Hybrid Sort

Whereas, *Hybrid Sort* is (combination of Insertion Sort and Merge Sort).

The algorithms are tested on three different types of inputs: <span style="color: blue">*unsorted random elements*</span>, <span style="color: purple">*already sorted elements*</span>, and <span style="color: orange">*mostly sorted elements*.</span> The time taken to sort various sizes of inputs is measured in seconds

### COMPARING ALGORITHMS
#### **Simple Sort**

Simple Sorting involves iterating through an array, comparing two adjacent elements, and swapping them if they are not in the correct order. We used the selection sort algorithm in this case. The following table shows the time taken to sort arrays of different sizes and degrees of sorting using Simple Sorting.

| --- | --- | --- | --- |
| 500 |	0.003722|	0.002444|	0.002310|
| 1000|	0.019431|	0.009164|	0.005980 |
|2000	|0.019349	|0.006842	|0.006063|
|5000|	0.066614	|0.034723	|0.031630|
|10000|	0.269310	|0.114728	|0.114622|

|200000	|132.959778	|44.635849|	42.926132|


From the table above, we can see that Simple Sorting is not very efficient for large arrays. It takes a lot of time to sort unsorted arrays, especially as the size of the array increases. The time taken to sort an already sorted or mostly sorted array is less than that taken to sort an unsorted array. This is because, in an already sorted array, the elements are already in the correct order, and hence, no swapping is necessary.

```python
import matplotlib.pyplot as plt

array_size = [500, 1000, 2000, 5000, 10000, 20000, 50000, 100000, 200000]
unsorted_time = [0.003722, 0.019431, 0.019349, 0.066614, 0.269310, 1.150233, 7.998028, 32.893887, 132.959778]
sorted_time = [0.002444, 0.009164, 0.006842, 0.034723, 0.114728, 0.429059, 2.866983, 11.750633, 44.635849]
mostly_sorted_time = [0.002310, 0.005980, 0.006063, 0.031630, 0.114622, 0.432852, 2.780592, 12.748405, 42.926132]

with plt.style.context('seaborn'):
    plt.plot(array_size, unsorted_time, label='Unsorted', color='blue', marker='o', linestyle='dashed', linewidth=2, markersize=6)
    plt.plot(array_size, sorted_time, label='Already Sorted', color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)
    plt.plot(array_size, mostly_sorted_time, label='Mostly Sorted', color='green', marker='o', linestyle='dashed', linewidth=2, markersize=6)

    plt.xlabel('Array Size')
    plt.ylabel('Time (seconds)')
    plt.title('Simple Sorting Algorithm')
    plt.legend()
    #Save the plot
    plt.savefig('simple_sorting.png')
    plt.show()
```

### **Insertion Sort**

Looking at the results of Insertion Sort, we can observe that its execution time increases as the size of the input array increases. Specifically, for unsorted arrays, the execution time of Insertion Sort scales quadratically with the input size. This is expected, as the worst-case time complexity of Insertion Sort is O(n^2). On the other hand, for arrays that are already sorted or mostly sorted, Insertion Sort performs significantly faster, with execution times that scale linearly with the input size. This is because the best-case time complexity of Insertion Sort is O(n), and when the array is already sorted or almost sorted, the algorithm can take advantage of this and skip some of the unnecessary comparisons and swaps.

### **Quick Sort**

We can also observe that the time taken to sort an array of mostly sorted elements is generally less than the time taken to sort an array of unsorted random elements of the same size. This is because the quick sort algorithm is designed to take advantage of existing orderings in the input array, and when an array is mostly sorted, it can take advantage of that existing ordering to reduce the number of comparisons and swaps needed.

Finally, we can see that the time taken to sort an already sorted array is less than the time taken to sort an unsorted random array of the same size. This is because the quick sort algorithm is designed to terminate early if it detects that the input array is already sorted, thereby avoiding unnecessary comparisons and swaps.

### **Merge Sort**

Looking at the results for Merge Sort, we can see that it performs better than Quick Sort in terms of worst-case time complexity. As the size of the input array increases, the time taken to sort the array also increases, but at a slower rate compared to Quick Sort.

In the case of unsorted random elements, Merge Sort performs slower than Quick Sort for smaller array sizes, but as the size of the array increases, Merge Sort becomes faster than Quick Sort. This is because the worst-case time complexity of Merge Sort is O(nlogn), whereas the worst-case time complexity of Quick Sort is O(n^2).

When the input array is already sorted, Merge Sort performs slower than Quick Sort for smaller array sizes, but as the size of the array increases, Merge Sort becomes faster than Quick Sort. This is because Merge Sort still has to perform the merging operation, whereas Quick Sort can skip some of its operations since the array is already sorted.

When the input array is mostly sorted, Merge Sort performs similarly to Quick Sort, with some variations depending on the size of the array. In general, Merge Sort's worst-case time complexity of O(nlogn) makes it a good choice for sorting arrays of any kind, regardless of their initial order.

On the other hand, merge sort has a more predictable and consistent performance for all input sizes, due to its divide-and-conquer approach, which can efficiently handle larger input sizes.

In addition, the performance of each algorithm can also be affected by the initial order of the input array. For example, if the input array is already sorted or mostly sorted, then the merge sort algorithm will be faster than the hybrid sort, since it can take advantage of the initial order and avoid unnecessary comparisons and swaps.

## **Performance**

Performance is checked by  *perf stat ./executable* .

The text file is available Reports/perf.txt .

Based on the performance counter stats provided, it seems that the Merge Sort algorithm has the best performance with the lowest task-clock time, the lowest number of cycles, and the lowest number of instructions per cycle. It also has a relatively low number of page faults, branches, and branch misses.

The Hybrid Sort algorithm has the second-best performance, with a slightly higher task-clock time, cycles, instructions per cycle, and branch misses compared to Merge Sort. However, it still outperforms Insertion Sort and Quick Sort by a significant margin.

Insertion Sort has the worst performance, with the highest task-clock time, cycles, instructions per cycle, and branch misses among all the algorithms. It also has a high number of page faults and branches.

Quick Sort has a better performance compared to Insertion Sort, but still not as good as Merge Sort or Hybrid Sort.
